```python
# module imports
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
import seaborn as sns
```

```python
# read in long format data 
wide_league_data = '../data/processed/all_leagues_wide.csv' 

# Read in data
wide_data = pd.read_csv(wide_league_data)

# create DF
wide_league_data_df = pd.DataFrame(wide_data)

# show DF
wide_league_data_df
```

```python
# read in long format data 
league_data = '../data/processed/all_leagues_long.csv' 

# Read in data
long_data = pd.read_csv(league_data)

# create DF
long_df = pd.DataFrame(long_data)

# show DF
long_df
```

```python
# Display information about long format data
print(long_df.info())

# Display dimensions of dataframe
print(f'These are the dimensions of the DataFrame, in (rows,columns) format: {long_df.shape}')
```

```python
# Create descriptive statistics for numerical columns
long_df.describe()
```

```python
# Display amount of missing values
long_df.isnull().sum()
```

```python
# Find max and min values

max_salary =  long_df['Salary'].max()
min_salary = long_df['Salary'].min()

# Output target salaries
print(f'The highest salary in the dataset is {max_salary}')
print(f'The lowest salary in the dataset is {min_salary}')
```

```python
# Info on player with lowest salary
min_row = long_df['Salary'].idxmin()
min_full_row = long_df.iloc[min_row]

# Info on player with largest salary
max_row = long_df['Salary'].idxmax()
max_full_row = long_df.iloc[max_row]

print(min_full_row)
print(max_full_row)
```

## NOTE regarding missing values
As my data represents professional athlete salaries, spanning the years (at the largest range) 1991-2017, missing values are to be expected. As the average career of a professional athlete is under 10 years, the vast majority of the athletes in my datasets will have missing salary values for several years, either before they entered their respective league or for after their career ended.

```python
# Define a custom formatting function
def format_ticks(value, pos):
    return '{:.0f}M'.format(value / 1000000)
```

```python
# Set X-axis tick format 
plt.gca().xaxis.set_major_formatter(FuncFormatter(format_ticks))

# Create histogram of Salary data
plt.hist(long_df['Salary'], bins=20)
plt.xlabel('Salaries')
plt.ylabel('Frequency of Salary')
#plt.ticklabel_format(style='scientific', axis='x')

plt.title('Histogram of Pro Athlete Salary Data 1991-2017')
plt.show()
```

```python
# Box Plot to investigate value spread
sns.boxplot(x='Year', y='Salary', data=long_df)
plt.xlabel('Year')
plt.ylabel('Salary')
plt.title('Box Plot of Salaries by Year')
plt.xticks(rotation=75)
# Format yticks 
plt.gca().yaxis.set_major_formatter(FuncFormatter(format_ticks))
plt.show()
```

```python
# VISUALIZE DATA
# set x and y values from data
years = long_df['Year']
salaries = long_df['Salary']

# Create line plot
plt.plot(years, salaries)
plt.xlabel('Year')
plt.ylabel('Salaries($)')
plt.ticklabel_format(style='plain', axis='y')
plt.title('Major 4 Sports Salary Progression 1991-2017')
plt.show()
```

```python
# Correlation Heatmap
corr = long_df.corr()

sns.heatmap(corr, annot=True,cmap='coolwarm', fmt='.2f')
plt.title('Variable Correlation Heatmap')
plt.show()
```

```python
# Line plot visualizing the "Top 4 Major Sports" average salaries by year

# Group the data by the 'League' column
year_salary_groups = long_df.groupby('Year')['Salary'].mean()

# Plot the average salary for each league
year_salary_groups.plot(marker='o')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Average Salary')
plt.title('Average Salary Growth by Year')
plt.gca().yaxis.set_major_formatter(FuncFormatter(format_ticks))


# Show plot
plt.show()
```

```python
# Line plot, visualizing the average salary per year for each league

# Group the data by the 'League' column
grouped = long_df.groupby(['Year', 'League'])['Salary'].mean().unstack()

# Plot the average salary for each league
grouped.plot(marker='o')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Average Salary')
plt.title('Average Salary Trends by League')
plt.gca().yaxis.set_major_formatter(FuncFormatter(format_ticks))
plt.legend(title='League')

# Show plot
plt.show()
```

```python
# Line plot showing the Max Salary per Year by League

# Group the data by the 'League' column
grouped = long_df.groupby(['Year', 'League'])['Salary'].max().unstack()

# Plot the average salary for each league
grouped.plot(marker='o')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Max Salary')
plt.title('Maximum Salary Trends by League')
plt.gca().yaxis.set_major_formatter(FuncFormatter(format_ticks))
plt.legend(title='League')

# Show plot
plt.show()
```

### Michael Jordan Effect
We can see the large spike in Max salary in the NBA in the late 90's, this was caused by Michael Jordan signing his last contract with the Bulls. Being by far the best player in the league and coming off a 3-peat, 2 year hiatus. 

## Team Data

```python
# Read in Team Data
team_file = '../data/processed/teams_long.csv'
data = pd.read_csv(team_file, dtype= {'Year':'Int64', 'Player Payroll':'Int64'})

team_data_df = pd.DataFrame(data)
team_data_df
```

```python
# Review DataFrame info
team_data_df.info()
print(f'The shape of the data is:{team_data_df.shape}')
```

```python
# Summary Stats of data
team_data_df.describe()
```

```python
# Histogram to visualize data distribution
plt.hist(team_data_df['Player Payroll'], bins=20)
plt.xlabel('Player Payroll')
plt.ylabel('Data Frequency')
plt.title('Player Payroll Data Distribution 1991-2017')
plt.show()
```

```python
# Need to convert <NA> null values to NaN to be handleable by matplotlib
team_data_df = team_data_df.replace(to_replace=<NA>, np.nan)

team_data_df
```

```python
# Need to convert <NA> null values to NaN to be handleable by matplotlib
team_data_df["Player Payroll"] = team_data_df["Player Payroll"].replace('<NA>', np.nan)

# Re-run of histogram after null value conversion
plt.hist(team_data_df['Player Payroll'], bins=20)
plt.xlabel('Player Payroll')
plt.ylabel('Data Frequency')
plt.title('Player Payroll Data Distribution 1991-2017')
plt.show()
```

```python
# Columns Correlation 
corr = team_data_df.corr(numeric_only=True)
corr
```

```python
# Correlation Heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation between Year & Player Payroll')
plt.show()
```
